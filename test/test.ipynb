```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/smartphone.csv')
df.head()
df.shape
```

```python
df['price'] = df['price'].apply(lambda x: round(x*0.012, 2))
```

```python
counts = df['brand_name'].value_counts()
counts = counts[counts > 3]
include = counts.index.to_numpy()
```

```python
df = df[df['brand_name'].isin(include)]
df
```

```python
df.to_csv('smartphone_cleaned.csv')
```
